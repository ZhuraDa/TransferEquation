```python
import numpy as np
import matplotlib.pyplot as plt
import math as m
```

**Численное и аналитическое решение однородной системы**

```python
file1 = open("data_zero.txt", "r")
t_interest = 0
u_calc,u_anal,v_calc,v_anal, X = [],[],[],[],[]

while True:
    # считываем строку
    line = file1.readline()
  
    # прерываем цикл, если строка пустая
    if not line:
        break
    
    data = line.split(sep=" ", maxsplit=-1)
    if len(data)!=1:
      u_calc.append(float(data[0]))
      u_anal.append(float(data[1]))
      v_calc.append(float(data[2]))
      v_anal.append(float(data[3]))
      X.append(float(data[4]))
    else:
      t_interest=float(data[0])
# закрываем файл
file1.close

fig, ax = plt.subplots(2,1)
ax[0].plot(X,u_anal, '-', label="Аналитическое решение")
ax[0].plot(X,u_calc, 'o', label = "Численное решение")
ax[1].plot(X,v_anal, '-', label="Аналитическое решение")
ax[1].plot(X,v_calc, 'o', label = "Численное решение")
ax[0].set_title("u")
ax[1].set_title("v")
for _ in ax:
    _.legend()
plt.show()

```
